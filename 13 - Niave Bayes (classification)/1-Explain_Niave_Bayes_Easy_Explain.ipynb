### **Play Tennis Example**

- **Content**: 
  - A dataset of training examples for a "Play Tennis" decision-making scenario.
  
- **Details**:
  - Each row represents a day with specific weather conditions:
    - **Outlook**: Sunny, Overcast, Rain
    - **Temperature**: Hot, Mild, Cool
    - **Humidity**: High, Normal
    - **Wind**: Weak, Strong
  - The last column, **PlayTennis**, indicates whether tennis was played (Yes/No).
  
- **Purpose**:
  - Used to train a machine learning model to predict whether tennis will be played based on weather conditions.

---






### ** Calculating Probabilities**

- **Content**: 
  - Probabilities calculated during the learning phase of a Naive Bayes classifier.
  
- **Details**:
  - Probabilities are computed for each attribute (**Outlook**, **Temperature**, **Humidity**, **Wind**) given the class labels (**Play=Yes** or **Play=No**).
  - Example probabilities:
    - **Outlook=Sunny**:
      - P(Sunny|Yes) = 2/9
      - P(Sunny|No) = 3/5
    - **Temperature=Cool**:
      - P(Cool|Yes) = 3/9
      - P(Cool|No) = 1/5
  - Prior probabilities:
    - P(Play=Yes) = 9/14
    - P(Play=No) = 5/14
  
- **Purpose**:
  - These probabilities are used to predict the likelihood of playing tennis based on new weather conditions.

---






### **Test Phase**

- **Content**: 
  - The test phase of the Naive Bayes classifier, where a new instance is classified.
  
- **Details**:
  - **New Instance**:
    - Outlook = Sunny
    - Temperature = Cool
    - Humidity = High
    - Wind = Strong
  - Probabilities from the learning phase are used to calculate:
    - P(Yes|x) = [P(Sunny|Yes) * P(Cool|Yes) * P(High|Yes) * P(Strong|Yes)] * P(Play=Yes) = 0.0053
    - P(No|x) = [P(Sunny|No) * P(Cool|No) * P(High|No) * P(Strong|No)] * P(Play=No) = 0.0206
  - **Decision**:
    - Since P(Yes|x) < P(No|x), the classifier predicts **"No"** (tennis will not be played).
  
- **Purpose**:
  - Demonstrates how the classifier uses learned probabilities to make predictions on new data.

### **Flow Chart**

- **Content**: 
  - A flowchart describing the Naive Bayes classification process.
  
- **Steps**:
  1. **Start**: Begin the classification process.
  2. **For each attribute**:
     - Traverse the attribute list.
     - Find the probability of the attribute value belonging to a class.
     - Update the class probabilities based on the attribute value.
  3. **Check**:
     - If all attribute values have been examined, move to the next step.
     - If not, continue to the next attribute.
  4. **End**:
     - Once all attributes are processed, the final classification is made based on the computed probabilities.
  
- **Purpose**:
  - Provides a visual representation of the step-by-step process used by the Naive Bayes algorithm to classify new instances.
